```python
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib

matplotlib.rc("font", family='YouYuan')
```

## 简单的折线图

```python
# 创建画布
plt.figure(figsize=(20, 8), dpi=100)
row = [1, 2, 3, 4, 5, 6, 7]
column = [17, 17, 18, 15, 11, 11, 13]
plt.plot(row, column)
# 保存图像要在显示图像之前
plt.savefig('./temp/output/01_test.png')
plt.show()
plt.close()
```

## 折线图完善

```python
# 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
minute_list_beijing = [random.uniform(15, 18) for x in range(60)]
plt.figure(figsize=(20, 8))
plt.plot(range(60), minute_list_beijing)

# 描述信息
plt.xlabel('时间', fontsize=20)
plt.ylabel('温度', fontsize=20)
plt.title('温度变化折线图', fontsize=25)

# 坐标刻度
plt.xticks(range(0, 60, 5), ['11点{}分'.format(x) for x in range(0, 60, 5)])
plt.yticks(range(-30, 41, 5))

# 添加网格线
plt.grid(linestyle='--', alpha=1)
plt.show()
plt.close()
```

## 同一个坐标系绘制两根折线

```python
# 再添加一个城市的温度变化
minute_list_beijing = [random.uniform(15, 18) for _ in range(60)]
minute_list_shanghai = [random.uniform(0, 20) for _ in range(60)]
plt.figure(figsize=(20, 8))
plt.plot(range(60), minute_list_beijing, label='北京')
plt.plot(range(60), minute_list_shanghai, label='上海')

# 描述信息
plt.xlabel('时间', fontsize=20)
plt.ylabel('温度', fontsize=20)
plt.title('温度变化折线图', fontsize=25)

# 坐标刻度
plt.xticks(range(0, 60, 5), ['11点{}分'.format(x) for x in range(0, 60, 5)])
plt.yticks(range(-30, 41, 5))

# 添加网格线
plt.grid(linestyle='--', alpha=1)
# 添加图例
plt.legend(loc='upper right')
plt.show()
plt.close()
```

## 多个坐标系绘图

```python
# 创建画布 nrows 绘图区域的行数, ncols 绘图区域的列数
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

ax[0].plot(minute_list_shanghai, label='上海')
ax[1].plot(minute_list_beijing, label='北京', color='r')

# 描述信息
ax[0].set_xlabel('时间', fontsize=20)
ax[0].set_ylabel('温度', fontsize=20)
ax[0].set_title('温度变化折线图', fontsize=25)
ax[1].set_xlabel('时间', fontsize=20)
ax[1].set_ylabel('温度', fontsize=20)
ax[1].set_title('温度变化折线图', fontsize=25)

# 添加坐标轴刻度
ax[0].set_xticks(range(0, 60, 5))
ax[0].set_xticklabels(['11点{}分'.format(x) for x in range(0, 60, 5)])
ax[0].set_yticks(range(0, 41, 5))
ax[1].set_xticks(range(0, 60, 5))
ax[1].set_xticklabels(['11点{}分'.format(x) for x in range(0, 60, 5)])
ax[1].set_yticks(range(0, 41, 5))

ax[0].grid(linestyle='--', alpha=0.8)
ax[1].grid(linestyle='--', alpha=0.8)

ax[0].legend(loc='upper right')
ax[1].legend(loc='upper right')

plt.show()
plt.close()
```

## 绘制 sin 函数图像

```python
x_axis = np.linspace(-10, 10, 10000)
y_axis = np.sin(x_axis)

plt.figure(figsize=(20, 8))
plt.plot(x_axis, y_axis)
plt.show()
plt.close()
```
