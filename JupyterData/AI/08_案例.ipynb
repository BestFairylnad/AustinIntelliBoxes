## 生成数据，绘制出正态分布的左半部分数据的直方图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 创建数据
create_data = np.random.standard_normal(100000)
# 截取小于平均值的数据
create_data = create_data[create_data < np.mean(create_data)]

create_data
```

```python
  # 构图
plt.figure(figsize=(20, 8), dpi=100)
plt.hist(create_data, bins=1000)
plt.show()
plt.close()
```

## axes绘图区域的命名

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

fig.suptitle('PIC')

ax[0].set_title('1')
ax[1].set_title('2')

plt.show()
plt.close()
```
