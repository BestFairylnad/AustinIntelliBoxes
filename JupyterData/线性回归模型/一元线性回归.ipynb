# 一元线性回归

## 一元线性回归的代码实现

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 散点图
x = [[1], [2], [4], [5]]
y = [2, 4, 6, 8]
plt.scatter(x, y)
plt.show()
plt.close()
```

```python
regr = LinearRegression()
regr.fit(x, y)
```

```python
# 模型预测
y = regr.predict([[1.5], [2.5], [3.5], [4.5]])
plt.scatter(x, y)
plt.plot(x, regr.predict(x))
plt.show()
plt.close()
```

```python
# 线性回归方程构造
print('系数为: {}'.format(str(regr.coef_[0])))
print('截距为: {}'.format(str(regr.intercept_)))
```
